```python
df = pd.DataFrame({
    'const': 17,
    'increment': np.arange(1000),
    'random': np.random.rand(1000),
    'normal': np.append(np.random.normal(5, 2, 500), np.random.normal(20, 2, 500))
})
```

```python
df.sample(5)
```

```python
df.plot()
```

```python
df.random.plot()
```

```python
df.random.hist()
```

```python
df.normal.hist(bins=30)
```

```python
df.increment.plot(color='tomato')
df.normal.plot(color='steelblue', secondary_y=True)
```
