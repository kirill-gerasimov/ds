```python
from gensim.models import word2vec
import logging
logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)
```

```python
sentences = word2vec.Text8Corpus('text8')
print(sentences)
```

```python
features = 200
```

```python
model = word2vec.Word2Vec(sentences, size=features, iter=30)
```

```python
model.most_similar(positive=['woman', 'king'], negative=['man'], topn=1)
```

```python
model.most_similar(positive=['germany', 'moscow'], negative=['russia'], topn=4)
```

```python
import numpy as np
from numpy import *
words = "queen king he she dog cat west east north south chemical".split()
fulllvectors = np.zeros(shape=(len(words), features)).astype(float32)
for idx,word in enumerate(words) :
    fulllvectors[idx, :] = model[word]

print("vectors shape: ", fulllvectors.shape)
    
```

```python
from sklearn.manifold import TSNE

modelTsne = TSNE(n_components=2, random_state=0)
np.set_printoptions(suppress=True)
vectors = modelTsne.fit_transform(fulllvectors) 

from sklearn import preprocessing

#normalizer = preprocessing.Normalizer()
#vectors =  normalizer.fit_transform(vectors, 'l2')

print(vectors)

```

```python
import matplotlib.pyplot as plt


fig, ax = plt.subplots()
print(words)
print(vectors)
print("type vectors ", type(vectors))
print("type vectors shape", vectors.shape)

axis1 = vectors[:,0]
axis2 = vectors[:,1]
print("types axis1:",type(axis1))

for idx, word in enumerate(words):
    print(word, vectors[idx, 0], vectors[idx, 1])
    ax.annotate(word, (vectors[idx, 0],vectors[idx, 1] ))


plt.plot(axis1, axis2, linestyle='', marker='.')
plt.show()
```
